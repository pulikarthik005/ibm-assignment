```python
import pandas as pd
from bs4 import BeautifulSoup
import yfinance as yf
import plotly.express as px
import requests

```

```python
game_stop = yf.Ticker("GME")
```

```python
gme_data = game_stop.history(period="max")
```

```python
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
fig = px.line(gme_data, x='Date', y='Close', title='GameStop Closing Prices Over Time')
fig.show()

# Plotting the volume over time
fig = px.bar(gme_data, x='Date', y='Volume', title='GameStop Trading Volume Over Time')
fig.show()
```
